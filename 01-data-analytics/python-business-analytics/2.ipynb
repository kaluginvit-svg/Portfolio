```python
# Перечень задач

## 1. Уравнение для расчёта ежедневной нормы калорий
- **а.** Задать переменные вес, рост и возраст и присвоить им значения
- **b.** Написать формулу для переменных:
  - **i.** для женщин: (вес (кг) × 10) + (рост (см) × 6,25) − (возраст (лет) × 5) − 161
  - **ii.** для мужчин: (вес (кг) × 10) + (рост (см) × 6,25) − (возраст (лет) × 5) + 5

## 2. Задать строку "123.456"
- **а.** Преобразовать к `int`
- **b.** Преобразовать к `float`
- **с.** Есть ли отличие в результате?

## 3. Создать текстовую переменную «Мир Труд Май»
- **а.** Извлечь слово «Труд». Сохранить в переменную `a`
- **b.** Извлечь слово «риМ». Сохранить в переменную `b`
- **с.** Преобразовать `b` в «Рим»
- **d.** Сконкатенировать переменную `a` и строку «ный». Перезаписать в переменную `a`
- **е.** Сконкатенировать переменную `a`, " ", переменную `b`. Вывести результат.

## 4. Создать переменные
- **а.** `name` = ваше имя
- **b.** `game` = название вашей любимой игры
- **с.** `hours` = рекордное количество проведённых часов в игре в течение суток
- **d.** Вывести f-строку: фраза типа «Меня зовут Олег, однажды я проиграл в Мир Танков 12 часов»

## 5. Строка «Иванов Иван Иванович»
Найти индексы пробелов в строке, привязавшись к позициям вырезать имя, фамилию, отчество в отдельные переменные `first_name`, `last_name`, `middle_name`.

## 6. Фраза «Королева кавалеру подарила каравеллу»
- **а.** Подсчитать количество букв «а»
- **b.** Подсчитать количество букв «к» (всех регистров). Привести фразу к строчным буквам

## 7. Разделить строку на слова по пробелам «Мы пойдем другим путем»
- **а.** Использовать `str.split()`
- **b.** Использовать `import re`
```

```python
# Задание 1: ежедневная норма калорий
вес = 70    # кг
рост = 175  # см
возраст = 30  # лет

# для женщин
калории_жен = (вес * 10) + (рост * 6.25) - (возраст * 5) - 161
# для мужчин
калории_муж = (вес * 10) + (рост * 6.25) - (возраст * 5) + 5

print("Норма калорий (женщины):", калории_жен)
print("Норма калорий (мужчины):", калории_муж)
```

```python
# Задание 2: строка "123.456" → int и float
s = "123.456"

# а) к int (отбрасывается дробная часть, либо через float)
a_int = int(float(s))

# b) к float
b_float = float(s)

# c) отличие: int даёт целое 123, float — 123.456
print("int:", a_int)
print("float:", b_float)
print("Отличие: int — целое число, float — с дробной частью")
```

```python
# Задание 3: «Мир Труд Май»
текст = "Мир Труд Май"

# а) извлечь «Труд»
a = текст[4:8]   # или: текст.split()[1]

# b) извлечь «риМ» (слово «Мир» в обратном порядке)
b = текст[0:3][::-1]   # «Мир» → «риМ»

# c) преобразовать b в «Рим»
b = "Рим"

# d) a + "ный" → перезаписать a
a = a + "ный"

# e) сконкатенировать a, " ", b и вывести
результат = a + " " + b
print(результат)
```

```python
# Задание 4: name, game, hours и f-строка
name = "Виталий"
game = "Counter Strike"
hours = 12

print(f"Меня зовут {name}, однажды я проиграл в {game} {hours} часов")
```

```python
# Задание 5: «Иванов Иван Иванович» — имя, фамилия, отчество
s = "Иванов Иван Иванович"
p1 = s.index(" ")       # первый пробел
p2 = s.index(" ", p1 + 1)  # второй пробел

last_name = s[:p1]
first_name = s[p1 + 1:p2]
middle_name = s[p2 + 1:]

print("Фамилия:", last_name)
print("Имя:", first_name)
print("Отчество:", middle_name)
```

```python
# Задание 6: «Королева кавалеру подарила каравеллу»
фраза = "Королева кавалеру подарила каравеллу"

# а) количество букв «а»
count_a = фраза.count("а")
print("Букв 'а':", count_a)

# b) количество букв «к» (все регистры) — приводим к нижнему
фраза_ниж = фраза.lower()
count_k = фраза_ниж.count("к")
print("Букв 'к' (в нижнем регистре):", count_k)
print("Фраза в нижнем регистре:", фраза_ниж)
```

```python
# Задание 7: разделить «Мы пойдем другим путем» по пробелам
строка = "Мы пойдем другим путем"

# а) str.split()
слова_split = строка.split()
print("Через str.split():", слова_split)

# b) import re
import re
слова_re = re.split(r"\s+", строка)
print("Через re.split():", слова_re)
```
