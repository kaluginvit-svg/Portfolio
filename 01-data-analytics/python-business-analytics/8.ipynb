# Задание: students_exams.xlsx

1. Загрузить все листы книги students_exams.xlsx в одноимённые датафреймы  
2. Объединить results и new_results в results_total  
3. Джойн: results_total + exams по blanc_id; присоединить students по student_id  
4. Группировка по предмету — кол-во проведённых экзаменов (blanc_id)  
5. Группировка по дате — кол-во проведённых экзаменов (blanc_id)  
6. Группировка по предмету и дате — кол-во проведённых экзаменов (blanc_id)  
7. Группировка по имени студента — сумма баллов по всем предметам  
8. Группировка по городу — средний балл по Физике  
9. Сводная: exam_date, предмет, кол-во студентов, сдавших экзамен  
10. Сводная: check_date, inspector, средний балл result  
11. Для students — melt: city и district из столбцов в значения (длинный формат)

```python
# Задания 1–3: загрузка, results_total, merged (запустите эту ячейку первой)
import pandas as pd

xlsx = "students_exams.xlsx"
exams = pd.read_excel(xlsx, sheet_name="Exams")
results = pd.read_excel(xlsx, sheet_name="Results")
new_results = pd.read_excel(xlsx, sheet_name="New Results")
students = pd.read_excel(xlsx, sheet_name="Students")

print("Exams:", exams.shape, "Results:", results.shape, "New Results:", new_results.shape, "Students:", students.shape)

# Задание 2: объединить results и new_results
results_total = pd.concat([results, new_results], ignore_index=True)

# Задание 3: джойн results_total + exams, затем students
merged = results_total.merge(exams, on="blanc_id", how="inner")
merged = merged.merge(students, on="student_id", how="left")
print("merged:", merged.shape)
```

```python
# Просмотр results_total (создаётся в ячейке выше)
results_total
```

```python
# Просмотр merged (создаётся в первой ячейке)
merged
```

```python
# Задание 4: группировка по предмету — кол-во проведённых экзаменов (blanc_id)
merged.groupby("subject")["blanc_id"].count()
```

```python
# Задание 5: группировка по дате — кол-во проведённых экзаменов (blanc_id)
merged.groupby("exam_date")["blanc_id"].count()
```

```python
# Задание 6: группировка по предмету и дате — кол-во проведённых экзаменов (blanc_id)
merged.groupby(["subject", "exam_date"])["blanc_id"].count()
```

```python
# Задание 7: группировка по имени студента — сумма баллов по всем предметам
merged.groupby("full_name")["result"].sum()
```

```python
# Задание 8: группировка по городу — средний балл по Физике
physics = merged[merged["subject"] == "Физика"]
physics.groupby("city")["result"].mean()
```

```python
# Задание 9: сводная — exam_date, предмет, кол-во студентов, сдавших экзамен
pivot9 = merged.groupby(["exam_date", "subject"])["blanc_id"].count().reset_index()
pivot9.columns = ["exam_date", "subject", "students_count"]
pivot9
```

```python
# Задание 10: сводная — check_date, inspector, средний балл result
pivot10 = merged.groupby(["check_date", "inspector"])["result"].mean().reset_index()
pivot10.columns = ["check_date", "inspector", "avg_result"]
pivot10
```

```python
# Задание 11: melt для students — city и district из столбцов в значения (длинная таблица)
students_long = students.melt(
    id_vars=["student_id", "full_name", "birth_date", "school_number", "average_grade"],
    value_vars=["city", "district"],
    var_name="location_type",
    value_name="location_value"
)
students_long
```
