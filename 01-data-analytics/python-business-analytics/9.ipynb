# Задание: pharma_orders_nan.xlsx

1. Загрузить данные pharma_orders_nan.xlsx  
2. Посчитать количество пропусков по столбцам  
3. Отфильтровать записи, в которых есть пропуски  
4. Удалить строки с пропусками в поле drug  
5. Заменить пропуски в колонке price на среднюю цену по каждому лекарству (drug) в городе (city)  
6. Посчитать количество дублирующих записей  
7. Удалить явные дубликаты (drop_duplicates)  
8. Посмотреть уникальные значения по столбцу с названием аптек  
9. Заменить неявные дубликаты: Горздрав → Аптека Горздрав, Здравсити → Аптека Здравсити

```python
# Задание 1: загрузка данных (файл pharma_orders_nan.xlsx — положите в папку с ноутбуком)
import pandas as pd

df = pd.read_excel("pharma_orders_nan.xlsx")
df
```

```python
# Задание 2: количество пропусков по столбцам
df.isna().sum()
```

```python
# Задание 3: записи, в которых есть хотя бы один пропуск
df[df.isna().any(axis=1)]
```

```python
# Задание 4: удалить строки с пропусками в поле drug
df = df.dropna(subset=["drug"])
df
```

```python
# Задание 5: заменить пропуски в price на среднюю цену по (drug, city)
df["price"] = df["price"].fillna(df.groupby(["drug", "city"])["price"].transform("mean"))
df
```

```python
# Задание 6: количество дублирующих записей
df.duplicated().sum()
```

```python
# Задание 7: удалить явные дубликаты
df = df.drop_duplicates()
df
```

```python
# Задание 8: уникальные значения по столбцу с названием аптек
pharmacy_col = "pharmacy_name"
df[pharmacy_col].unique()
```

```python
# Задание 9: заменить неявные дубликаты (Горздрав → Аптека Горздрав, Здравсити → Аптека Здравсити)
df["pharmacy_name"] = df["pharmacy_name"].replace({"Горздрав": "Аптека Горздрав", "Здравсити": "Аптека Здравсити"})
df["pharmacy_name"].unique()
```
