# Задание: функции

1. Функция: строка → количество слов в ней  
2. Рекурсивная функция для n-го треугольного числа  
3. Функция: список чисел → список только положительных  
4. Функция: НОД двух чисел  
5. Функция: строка → заглавные↔строчные  
6. Функция: список строк → только строки длиной больше 5  
7. Рекурсивная функция: сумма цифр числа  
8. Функция: два списка → список общих элементов  
9. Функция: строка → количество цифр  
10. Функция: список чисел → True, если отсортирован по возрастанию  
11. Лямбда: максимальный элемент списка  
12. Лямбда: фильтр — только положительные числа

```python
# Задача 1: количество слов в строке
def count_words(s):
    return len(s.split())

print(count_words("один два три"))  # 3
```

```python
# Задача 2: рекурсия — n-е треугольное число (1+2+...+n)
def triangular(n):
    if n <= 0:
        return 0
    return n + triangular(n - 1)

print(triangular(5))  # 15
```

```python
# Задача 3: только положительные числа из списка
def positive_only(lst):
    return [x for x in lst if x > 0]

print(positive_only([1, -2, 3, -4, 5]))  # [1, 3, 5]
```

```python
# Задача 4: НОД двух чисел (алгоритм Евклида)
def gcd(a, b):
    a, b = abs(a), abs(b)
    while b:
        a, b = b, a % b
    return a

print(gcd(48, 18))  # 6
```

```python
# Задача 5: замена заглавных на строчные и наоборот
def swap_case(s):
    return s.swapcase()

print(swap_case("Привет Мир"))  # пРИВЕТ мИР
```

```python
# Задача 6: только строки длиной больше 5
def long_strings(lst):
    return [s for s in lst if len(s) > 5]

print(long_strings(["а", "привет", "мир", "программа"]))  # ['привет', 'программа']
```

```python
# Задача 7: рекурсия — сумма цифр числа
def sum_digits(n):
    n = abs(int(n))
    if n < 10:
        return n
    return n % 10 + sum_digits(n // 10)

print(sum_digits(12345))  # 15
```

```python
# Задача 8: элементы, которые есть в обоих списках
def common_elements(a, b):
    return list(set(a) & set(b))

print(common_elements([1, 2, 3], [2, 3, 4]))  # [2, 3] (порядок может отличаться)
```

```python
# Задача 9: количество цифр в строке
def count_digits(s):
    return sum(c.isdigit() for c in s)

print(count_digits("a1b22c"))  # 3
```

```python
# Задача 10: список отсортирован по возрастанию?
def is_sorted(lst):
    return lst == sorted(lst)

print(is_sorted([1, 2, 3]))   # True
print(is_sorted([1, 3, 2]))   # False
```

```python
# Задача 11: лямбда — максимальный элемент списка
max_elem = lambda lst: max(lst)
print(max_elem([3, 1, 4, 1, 5]))  # 5
```

```python
# Задача 12: лямбда — только положительные числа
positive = lambda lst: list(filter(lambda x: x > 0, lst))
print(positive([1, -2, 3, -4, 5]))  # [1, 3, 5]
```
