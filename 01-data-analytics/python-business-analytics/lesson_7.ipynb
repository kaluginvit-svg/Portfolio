2. Выделить Марс и Земля (loc)  
3. Планеты с диаметром > 10000 км (loc)  
4. Планеты: спутников > 3 и температура поверхности < 0  
5. Сортировка по возрастанию диаметра  
6. Максимальное количество спутников  
7. Минимальная масса  
8. Среднее (среднеарифметическое) диаметра  
9. Медиана диаметра  
10. Сумма масс всех планет  
11. Столбец «категория планеты» по массе: &lt;10 — Легкая, &lt;1000 — Средняя, ≥1000 — Тяжелая (apply)  
12. Выгрузка в xlsx

```python
# Задание 1: загрузка данных и обзор
import pandas as pd

df = pd.read_csv("astronomical_observations.csv", encoding="utf-8")
df_climate = pd.read_csv("climate.csv", encoding="utf-8")

# Основная таблица — планеты
print("=== info() ===")
df.info()
print("\n=== shape ===")
print(df.shape)
print("\n=== head() ===")
df.head()
```

```python
df.tail()
```

```python
# Задание 2: Марс и Земля — loc
df.loc[df["Название планеты"].isin(["Марс", "Земля"])]
```

```python
# Задание 3: планеты с диаметром более 10000 км — loc
df.loc[df["Диаметр (км)"] > 10000]
```

```python
# Задание 4: более 3 спутников и температура поверхности < 0
df.loc[(df["Количество спутников"] > 3) & (df["Температура поверхности (°C)"] < 0)]
```

```python
# Задание 5: сортировка по возрастанию диаметра
df_sorted = df.sort_values("Диаметр (км)")
df_sorted
```

```python
# Задание 6: максимальное количество спутников
df["Количество спутников"].max()
```

```python
# Задание 7: минимальная масса планеты
df["Масса (10^24 кг)"].min()
```

```python
# Задание 8: среднее (среднеарифметическое) диаметра
df["Диаметр (км)"].mean()
```

```python
# Задание 9: медиана диаметра
df["Диаметр (км)"].median()
```

```python
# Задание 10: сумма масс всех планет
df["Масса (10^24 кг)"].sum()
```

```python
# Задание 11: столбец «категория планеты» по массе (apply)
def category_by_mass(m):
    if m < 10:
        return "Легкая"
    if m < 1000:
        return "Средняя"
    return "Тяжелая"

df["Категория планеты"] = df["Масса (10^24 кг)"].apply(category_by_mass)
df
```

```python
# Задание 12: выгрузка в xlsx
df.to_excel("planets_result.xlsx", index=False)
print("Таблица сохранена в planets_result.xlsx")
```
