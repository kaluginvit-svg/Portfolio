# Задачи: множества, кортежи, списки

## Множества (задачи 1–10)
1. Создать множество из списка имён `['Alice', 'Bob', 'Charlie', 'Alice', 'David']`
2. Объединить два множества: `{'Alice', 'Bob'}` и `{'Charlie', 'David'}`
3. Найти пересечение: `{'Alice', 'Bob'}` и `{'Bob', 'Charlie'}`
4. Найти разность: `{'Alice', 'Bob', 'Charlie'}` и `{'Bob', 'David'}`
5. Найти симметрическую разность: `{'Alice', 'Bob'}` и `{'Bob', 'Charlie'}`
6. Добавить элемент `'Eve'` в множество `{'Alice', 'Bob', 'Charlie'}`
7. Удалить элемент `'Bob'` с помощью метода **discard**
8. Скопировать множество `{'Alice', 'Bob', 'Charlie'}`
9. Очистить множество `{'Alice', 'Bob', 'Charlie'}`
10. Удалить случайный элемент с помощью метода **pop**

## Методы кортежей (задачи 11–15)
11. Найти индекс первого вхождения `'Charlie'` в кортеже `('Alice', 'Bob', 'Charlie', 'Alice', 'David')`
12. Подсчитать количество элементов `'Alice'` в этом кортеже
13. Найти длину кортежа `('Alice', 'Bob', 'Charlie', 'David')`
14. Найти максимальное значение в этом кортеже
15. Проверить, есть ли хотя бы один истинный элемент в `('', '', 'Charlie', '')` с помощью **any**

## Методы списков (задачи 16–24)
16. Найти индекс первого вхождения `'Charlie'` в списке
17. Подсчитать количество элементов `'Alice'` в списке
18. Добавить `'Eve'` в конец списка
19. Удалить элемент `'Bob'` из списка
20. Перевернуть список с помощью **reverse**
21. Расширить список `['Alice', 'Bob']` элементами `['Charlie', 'David']` с помощью **extend**
22. Удалить последний элемент с помощью **pop**
23. Вставить `'Eve'` на индекс 2 с помощью **insert**
24. Отсортировать список `['David', 'Charlie', 'Bob', 'Alice']` с помощью **sort**

```python
# Задача 1: множество из списка имён
names_list = ['Alice', 'Bob', 'Charlie', 'Alice', 'David']
names_set = set(names_list)
print(names_set)
```

```python
# Задача 2: объединение двух множеств
a = {'Alice', 'Bob'}
b = {'Charlie', 'David'}
union = a | b   # или a.union(b)
print(union)
```

```python
# Задача 3: пересечение двух множеств
a = {'Alice', 'Bob'}
b = {'Bob', 'Charlie'}
intersection = a & b   # или a.intersection(b)
print(intersection)
```

```python
# Задача 4: разность множеств
a = {'Alice', 'Bob', 'Charlie'}
b = {'Bob', 'David'}
diff = a - b   # или a.difference(b)
print(diff)
```

```python
# Задача 5: симметрическая разность
a = {'Alice', 'Bob'}
b = {'Bob', 'Charlie'}
sym_diff = a ^ b   # или a.symmetric_difference(b)
print(sym_diff)
```

```python
# Задача 6: добавить элемент в множество
s = {'Alice', 'Bob', 'Charlie'}
s.add('Eve')
print(s)
```

```python
# Задача 7: удалить элемент с помощью discard
s = {'Alice', 'Bob', 'Charlie'}
s.discard('Bob')
print(s)
```

```python
# Задача 8: скопировать множество
s = {'Alice', 'Bob', 'Charlie'}
s_copy = s.copy()
print(s_copy)
```

```python
# Задача 9: очистить множество
s = {'Alice', 'Bob', 'Charlie'}
s.clear()
print(s)
```

```python
# Задача 10: удалить случайный элемент с помощью pop
s = {'Alice', 'Bob', 'Charlie'}
removed = s.pop()
print("Удалён:", removed)
print("Осталось:", s)
```

```python
# Задача 11: индекс первого вхождения 'Charlie' в кортеже
t = ('Alice', 'Bob', 'Charlie', 'Alice', 'David')
idx = t.index('Charlie')
print(idx)
```

```python
# Задача 12: количество элементов 'Alice' в кортеже
t = ('Alice', 'Bob', 'Charlie', 'Alice', 'David')
cnt = t.count('Alice')
print(cnt)
```

```python
# Задача 13: длина кортежа
t = ('Alice', 'Bob', 'Charlie', 'David')
print(len(t))
```

```python
# Задача 14: максимальное значение в кортеже (лексикографически для строк)
t = ('Alice', 'Bob', 'Charlie', 'David')
print(max(t))
```

```python
# Задача 15: any — есть ли хотя бы один истинный элемент
t = ('', '', 'Charlie', '')
result = any(t)
print(result)
```

```python
# Задача 16: индекс первого вхождения 'Charlie' в списке
lst = ['Alice', 'Bob', 'Charlie', 'Alice', 'David']
idx = lst.index('Charlie')
print(idx)
```

```python
# Задача 17: количество элементов 'Alice' в списке
lst = ['Alice', 'Bob', 'Charlie', 'Alice', 'David']
cnt = lst.count('Alice')
print(cnt)
```

```python
# Задача 18: добавить 'Eve' в конец списка
lst = ['Alice', 'Bob', 'Charlie']
lst.append('Eve')
print(lst)
```

```python
# Задача 19: удалить элемент 'Bob' из списка
lst = ['Alice', 'Bob', 'Charlie']
lst.remove('Bob')
print(lst)
```

```python
# Задача 20: перевернуть список с помощью reverse
lst = ['Alice', 'Bob', 'Charlie']
lst.reverse()
print(lst)
```

```python
# Задача 21: расширить список с помощью extend
lst = ['Alice', 'Bob']
lst.extend(['Charlie', 'David'])
print(lst)
```

```python
# Задача 22: удалить последний элемент с помощью pop
lst = ['Alice', 'Bob', 'Charlie']
last = lst.pop()
print("Удалён:", last)
print("Список:", lst)
```

```python
# Задача 23: вставить 'Eve' на индекс 2 с помощью insert
lst = ['Alice', 'Bob', 'Charlie']
lst.insert(2, 'Eve')
print(lst)
```

```python
# Задача 24: отсортировать список в алфавитном порядке с помощью sort
lst = ['David', 'Charlie', 'Bob', 'Alice']
lst.sort()
print(lst)
```
