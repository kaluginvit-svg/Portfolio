```python
# Задача 1: словарь служб и телефонов
services = {
    "Пожарная служба": "101",
    "Полиция": "102",
    "Скорая помощь": "103",
    "Газовая служба": "104",
    "Служба спасения": "112"
}
print(services)
```

```python
# Задача 2: словарь через dict()
d = dict(Пожарная_служба="101", Полиция="102")
# или: d = dict([("Пожарная служба", "101"), ("Полиция", "102")])
d = dict([("Пожарная служба", "101"), ("Полиция", "102")])
print(d)
```

```python
# Задача 3: список кортежей в словарь
pairs = [("Пожарная служба", "101"), ("Полиция", "102"), ("Скорая помощь", "103")]
d = dict(pairs)
print(d)
```

```python
# Задача 4: добавить ключ "Газовая служба": "104"
d = {"Пожарная служба": "101", "Полиция": "102"}
d["Газовая служба"] = "104"
print(d)
```

```python
# Задача 5: обновить "Полиция" на "911"
d = {"Пожарная служба": "101", "Полиция": "102"}
d["Полиция"] = "911"
print(d)
```

```python
# Задача 6: удалить ключ "Скорая помощь"
d = {"Пожарная служба": "101", "Полиция": "102", "Скорая помощь": "103"}
del d["Скорая помощь"]
print(d)
```

```python
# Задача 7: update — добавить "Служба спасения": "112"
d = {"Пожарная служба": "101", "Полиция": "102"}
d.update({"Служба спасения": "112"})
print(d)
```

```python
# Задача 8: все ключи
d = {"Пожарная служба": "101", "Полиция": "102", "Газовая служба": "104"}
print(d.keys())
print(list(d.keys()))
```

```python
# Задача 9: все значения
d = {"Пожарная служба": "101", "Полиция": "102", "Газовая служба": "104"}
print(d.values())
print(list(d.values()))
```

```python
# Задача 10: все пары ключ-значение (items)
d = {"Пожарная служба": "101", "Полиция": "102", "Газовая служба": "104"}
print(d.items())
print(list(d.items()))
```

```python
# Задача 11: pop — удалить и вернуть значение по ключу "Газовая служба"
d = {"Пожарная служба": "101", "Полиция": "102", "Газовая служба": "104"}
value = d.pop("Газовая служба")
print("Удалено:", value)
print("Словарь:", d)
```

```python
# Задача 12: popitem — удалить и вернуть последнюю пару ключ-значение
d = {"Пожарная служба": "101", "Полиция": "102", "Газовая служба": "104"}
pair = d.popitem()
print("Удалено:", pair)
print("Словарь:", d)
```

```python
# Задача 13: get с значением по умолчанию
d = {"Пожарная служба": "101", "Полиция": "102"}
value = d.get("Скорая помощь", "Номер не найден")
print(value)
```

```python
# Задача 14: setdefault — добавить "Техническая поддержка": "105" только если ключа нет
d = {"Пожарная служба": "101", "Полиция": "102"}
d.setdefault("Техническая поддержка", "105")
print(d)
```

```python
# Задача 15: clear — очистить словарь
d = {"Пожарная служба": "101", "Полиция": "102", "Газовая служба": "104"}
d.clear()
print(d)
```

```python
# Задача 16: текущие дата и время — формат "Час:Минута:Секунда - День/Месяц/Год"
from datetime import datetime
now = datetime.now()
s = now.strftime("%H:%M:%S - %d/%m/%Y")
print(s)
```

```python
# Задача 17: объект datetime из строки "15-08-2020 10:30:00"
from datetime import datetime
s = "15-08-2020 10:30:00"
dt = datetime.strptime(s, "%d-%m-%Y %H:%M:%S")
print(dt)
```

```python
# Задача 18: текущая дата в формате "Год-Месяц-День"
from datetime import date
today = date.today()
print(today.strftime("%Y-%m-%d"))
```

```python
# Задача 19: текущая дата и время, время — полдень (12:00)
from datetime import datetime
now = datetime.now()
noon_today = now.replace(hour=12, minute=0, second=0, microsecond=0)
print(noon_today)
```

```python
# Задача 20: дата ровно через три месяца
from datetime import date
from calendar import monthrange
d = date.today()
m, y = d.month + 3, d.year
while m > 12:
    m -= 12
    y += 1
last_day = monthrange(y, m)[1]
day = min(d.day, last_day)
future = date(y, m, day)
print(future)
```

```python
# Задача 21: какой день недели через 30 дней
from datetime import date, timedelta
today = date.today()
future = today + timedelta(days=30)
weekdays = ["Пн", "Вт", "Ср", "Чт", "Пт", "Сб", "Вс"]
print(f"Через 30 дней: {future}, день недели: {weekdays[future.weekday()]}")
```
