# Задачи: условия и циклы

## Условия (задачи 1–6)
1. Число чётное или нечётное  
2. Три числа — определить, является ли одно из них средним (не макс. и не мин.)  
3. Число положительное и кратно 5  
4. Число отрицательное или делится на 3  
5. Ввод числа → вывод квадрата; при некорректном вводе — сообщение об ошибке  
6. Два числа: второе — делитель первого? Если второе = 0 — try/except, сообщение «Деление на 0 невозможно»  

## Циклы (задачи 7–14)
7. Числа от 1 до 100, которые делятся на 7  
8. Сумма чисел от 1 до 50  
9. Список слов от пользователя — вывод с индексами  
10. Список чисел от пользователя — вывод с индексами  
11. Запрашивать число, пока не введено положительное  
12. Запрашивать пароль, пока не введён правильный (например 12345)  
13. Числа от 1 до 50, пропуская делящиеся на 5  
14. Проверка: число положительное, чётное и меньше 50

```python
# Задача 1: чётное или нечётное
n = int(input("Введите число: "))
if n % 2 == 0:
    print("Чётное")
else:
    print("Нечётное")
```

```python
# Задача 2: одно из трёх чисел — среднее (не макс. и не мин.)
a = int(input("Число 1: "))
b = int(input("Число 2: "))
c = int(input("Число 3: "))
mn, mx = min(a, b, c), max(a, b, c)
for x in (a, b, c):
    if mn < x < mx:
        print(f"Среднее: {x}")
        break
else:
    print("Нет единственного среднего (есть равные)")
```

```python
# Задача 3: число положительное и кратно 5
n = int(input("Введите число: "))
if n > 0 and n % 5 == 0:
    print("Да, положительное и кратно 5")
else:
    print("Нет")
```

```python
# Задача 4: число отрицательное или делится на 3
n = int(input("Введите число: "))
if n < 0 or n % 3 == 0:
    print("Да")
else:
    print("Нет")
```

```python
# Задача 5: ввод числа → квадрат; при ошибке — сообщение
try:
    n = float(input("Введите число: "))
    print("Квадрат:", n ** 2)
except ValueError:
    print("Ошибка: введено не число")
```

```python
# Задача 6: второе число — делитель первого; при 0 — try/except
try:
    a = int(input("Первое число: "))
    b = int(input("Второе число: "))
    if b == 0:
        print("Деление на 0 невозможно")
    elif a % b == 0:
        print("Да, второе — делитель первого")
    else:
        print("Нет, не делитель")
except ValueError:
    print("Ошибка ввода")
```

```python
# Задача 7: числа от 1 до 100, делящиеся на 7
for i in range(1, 101):
    if i % 7 == 0:
        print(i, end=" ")
```

```python
# Задача 8: сумма чисел от 1 до 50
s = sum(range(1, 51))
print("Сумма:", s)
```

```python
# Задача 9: список слов — вывод с индексами
words = input("Введите слова через пробел: ").split()
for i, w in enumerate(words):
    print(i, w)
```

```python
# Задача 10: список чисел — вывод с индексами
raw = input("Введите числа через пробел: ").split()
numbers = [float(x) for x in raw]
for i, n in enumerate(numbers):
    print(i, n)
```

```python
# Задача 11: запрашивать число, пока не введено положительное
while True:
    n = float(input("Введите положительное число: "))
    if n > 0:
        print("Спасибо!")
        break
    print("Нужно положительное. Повторите.")
```

```python
# Задача 12: запрашивать пароль, пока не верный (например 12345)
password = "12345"
while True:
    p = input("Введите пароль: ")
    if p == password:
        print("Вход разрешён")
        break
    print("Неверный пароль")
```

```python
# Задача 13: числа от 1 до 50, пропуская делящиеся на 5
for i in range(1, 51):
    if i % 5 == 0:
        continue
    print(i, end=" ")
```

```python
# Задача 14: число положительное, чётное и меньше 50
n = int(input("Введите число: "))
if n > 0 and n % 2 == 0 and n < 50:
    print("Да")
else:
    print("Нет")
```
